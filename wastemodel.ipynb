```python
import tensorflow as tf
```

```python
IMG_size = 200
```

```python
data_generator = tf.keras.preprocessing.image.ImageDataGenerator(
    horizontal_flip = True,
    vertical_flip = True,
    rotation_range = 30)
```

```python
train_ds = data_generator.flow_from_directory(
    '/kaggle/input/waste-classification-data/DATASET/TRAIN/',
   target_size = (IMG_size,IMG_size),
   seed = 123,
)
```

```python
valid_ds = data_generator.flow_from_directory(
    '/kaggle/input/waste-classification-data/DATASET/TEST/',
   target_size = (IMG_size,IMG_size),
   seed = 123,
)
```

```python
InvModel = tf.keras.applications.InceptionV3(input_shape=(224,224,3),classes = 2,include_top = False,pooling = 'max')
print(InvModel.summary())
```

```python
model = tf.keras.models.Sequential(InvModel)
```

```python
model.add(tf.keras.layers.Dense(32,activation = 'softmax'))
model.add(tf.keras.layers.Dense(2,activation = 'softmax'))
model.summary()
```

```python
model.compile(
    optimizer='SGD',
    loss = 'mse',
    metrics = ['accuracy','AUC']
)
```

```python
model.fit_generator(
    train_ds,
    epochs = 25
)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,15))
plt.plot(model.history.history['loss'],marker = '*',linestyle = '--',markeredgecolor = 'red',color = 'green',linewidth = 2.5,markersize = 17)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Model Evaluation of Waste Model')
plt.grid()
plt.savefig('Model Evaluation of Waste Model - Loss.png')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,15))
plt.plot(model.history.history['accuracy'],marker = '*',linestyle = '--',markeredgecolor = 'red',color = 'green',linewidth = 2.5,markersize = 17)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Model Evaluation of Waste Model')
plt.grid()
plt.savefig('Model Evaluation of Waste Model - Acc.png')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,15))
plt.plot(model.history.history['auc'],marker = '*',linestyle = '--',markeredgecolor = 'red',color = 'green',linewidth = 2.5,markersize = 17)
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.title('Model Evaluation of Waste Model')
plt.grid()
plt.savefig('Model Evaluation of Waste Model - AUC.png')
```
